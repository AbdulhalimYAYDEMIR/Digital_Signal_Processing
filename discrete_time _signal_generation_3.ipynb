```python
'''
AbdÃ¼lhalim YAYDEMÄ°R  
'''
# DSP environment' da numpy ve matplotlib paketleri mevcut deÄŸil onlar anaconda Ã¼zerinden eklendi 
import numpy as np 
import matplotlib.pyplot as plt           
%matplotlib notebook                     

# font ayarÄ± https://matplotlib.org/stable/gallery/text_labels_and_annotations/text_fontdict.html#sphx-glr-gallery-text-labels-and-annotations-text-fontdict-py
font = {'family': 'serif',
       'color'  : 'black',
       'weight' : 'normal',
       'size'   :  18,
       }


# bÃ¶yle yaparak her seferinde np.pi demek yerine sadece pi diyebliriz                           
pi = np.pi 


# gÃ¶zlemlenecek aralÄ±ÄŸÄ± belirle
t = np.arange(0,100+0.001,pi/100)
n = np.arange(0,100,1) 


# x[n] = sin(pi*n/19)
# degiÅŸkene denklem atama      https://numpy.org/doc/stable/reference/generated/numpy.sin.html
continous = np.sin(0.1*pi*t)+(1/3)*np.sin(0.3*pi*t)+(1/5)*np.sin(0.4*pi*t)
discrete = np.sin(0.05*pi*n)+(1/3)*np.sin(0.15*pi*n)+(1/5)*np.sin(0.2*pi*n)


# sinyal Ã§izdirildi

plt.subplot(2,1,1)
plt.plot(t, continous,'b-')
plt.title('x(t)', fontdict=font)
plt.grid(True)
plt.annotate('T = 20s', xy=(21, 1.12), xytext=(-13, 1.16), arrowprops=dict(facecolor='red', shrink=0.05))


plt.subplot(2,1,2)
plt.stem(n, discrete,'r-')
plt.title('x[n]', fontdict=font)
plt.xlabel('t,n')
plt.grid(True)
plt.annotate('N = 40 Ã–RNEK', xy=(82, 1.18), xytext=(15, 1.20), arrowprops=dict(facecolor='blue', shrink=0.05))
plt.ylim(-1.3,1.4)

plt.show()
```
