### SORU-1

## a)

###         İşlemler
##### H(e^jw )=∑〖x[n] e^(-jwn) 〗
##### =∑〖(δ[n]-δ[n-1])e^(-jwn) 〗
##### =∑〖δ[n]e^(-jwn) 〗- ∑〖δ[n-1]e^(-jwn) 〗
##### =1-e^(-jw)
##### =2je^((-jw)/2) ((e^(jw/2)-e^((-jw)/2)))/2j
##### =2je^((-jw)/2) sin⁡(w/2)
##### =2e^(jπ/2) e^((-jw)/2) sin⁡(w/2)
##### =2e^(j((-w)/2+π/2)) sin⁡(w/2)

##                   Sonuç 
    
### Genlik:
### |H(e^jw)| = |2sin⁡(w/2)|

### Faz:
### ∡H(e^jw ) =  (-w)/2 + π/2           0≤w<2π


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
%matplotlib notebook

w_cont=np.arange(0,2*np.pi,np.pi/1000)

H_abs=np.abs(2*np.sin(w_cont/2))
H_phase=(-w_cont/2) + (np.pi/2)             #np.angle(np.exp(1j*((-w_cont/2)+(np.pi/2))))

plt.subplot(121)
plt.plot(w_cont/np.pi,H_abs)
plt.title('$|H(e^{j\omega})|$')
plt.xlabel('Normalize frekans (x$\pi$)')

plt.subplot(122)
plt.plot(w_cont/np.pi,H_phase/np.pi)
plt.title('$ \measuredangle H(e^{j\omega}) (\\times \pi) $')
plt.xlabel('Normalize frekans (x$\pi$)')
```
