```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib notebook


x = np.zeros(5)
y = np.zeros(5)


y[-1] = 0
x[0] = 1
x[1] = 0

for n in range(0,4):
    y[n] = 1.2*y[n-1]+x[n]-x[n-4]
    
print('y[n]in ilk dört değeri : ' ,y[0:4])

plt.figure()
n = np.arange(0,4)
plt.stem(n,y[0:4])
plt.title('y[n]=h[n]')
plt.xlabel('n')
plt.grid(True)
```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib notebook

h = np.array([1/3,1/3,1/3],dtype=float)
x = np.zeros(10)
for n in range(0,10):
    x[n] = np.cos(np.pi*(n+10)/5)

nHmin = -1
nXmin = 10
M = len(x)
N = len(h)
L = M + N -1
y = np.zeros(L)
for n in range(L):
    y[n] = 0
    for k in range (M):
        if (n-k)>=0 and (n-k)<N:
            y[n] = y[n]+x[k]*h[n-k]

            
            

n = np.arange(10,20)
nY = [i for i in range(nXmin+nHmin,nXmin+nHmin+L)]

print('\nx1[n] : \n')
print(x)
print('\nx1[n] indisleri : ',n)

print('\n\ny1[n] : \n')
print(y)
print('\ny1[n] indisleri : ',nY)



plt.figure()
plt.subplot(1,2,1)
plt.stem(n,x)
plt.title('x1[n]')
plt.xlabel('n')
plt.grid(True)

plt.subplot(1,2,2)
plt.stem(nY,y)
plt.title('y1[n]')
plt.xlabel('n')
plt.grid(True)
```

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib notebook

h = np.array([1/3,1/3,1/3],dtype=float)
x = np.array([(-0.3)**3,(-0.3)**4],dtype=float)

nHmin = -1
nXmin = 3
M = len(x)
N = len(h)
L = M + N -1
y = np.zeros(L)
for n in range(L):
    y[n] = 0
    for k in range (M):
        if (n-k)>=0 and (n-k)<N:
            y[n] = y[n]+x[k]*h[n-k]


n = np.arange(3,5)
nY = [i for i in range(nXmin+nHmin,nXmin+nHmin+L)]

print('x2[n] : ',x)
print('x2[n] indisleri : ',n)

print('\ny2[n] : ',y)
print('y2[n] indisleri : ',nY)

plt.figure()
plt.subplot(1,2,1)
plt.stem(n,x)
plt.title('x2[n]')
plt.xlabel('n')
plt.grid(True)

plt.subplot(1,2,2)
plt.stem(nY,y)
plt.title('y2[n]')
plt.xlabel('n')
plt.grid(True)
```
