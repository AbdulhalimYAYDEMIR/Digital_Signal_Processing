```python
'''
Abdülhalim YAYDEMİR
'''
import numpy as np
from matplotlib import pyplot as plt

pi=np.pi
w_cont=np.arange(-pi,pi,pi/1000)
H_phase=np.array([-w if pi*0.25<np.absolute(w)<=pi*0.5 else 0 for w in w_cont])
H_abs=np.array([2 if pi*0.25<np.absolute(w)<=pi*0.5 else 0 for w in w_cont])

plt.subplot(121)
plt.plot(w_cont/np.pi,H_abs)
plt.title('$|H(e^{j\omega})|$')
plt.xlabel('Normalize frekans (x$\pi$)')
plt.subplot(122)
plt.plot(w_cont/np.pi,H_phase/np.pi)
plt.title('$ \measuredangle H(e^{j\omega})$')
plt.xlabel('Normalize frekans (x$\pi$)')
```
