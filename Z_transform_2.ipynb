### SORU-2

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sgnl

n = np.arange(0,10,1)

# ELLE
h_n_elle =(-1/4)*((-1)**n)+(-1/4)*((1)**n)+(1/2)*((1)**(n+1))*(n+1)

# RESİDUEZ KOMUTU (ÖRNEK-7 PDF) İLE
z = np.array([0])
p = np.array([-1,1,1])
b,a = sgnl.zpk2tf(z,p,1)
r,p,k =sgnl.residuez(b,a) #residu +
poless = np.roots(a)
print('poless:',poless)
print('r     :',r)
print('p     :',p)
print('k     :',k)
h_n_residuez =(1/4)*((-1)**n)+(1/4)*((1)**n)+(1/2)*((1)**(n+1))*(n+1)

# DİMPULSE KOMUTU(ÖRNEK-8 PDF) İLE
num = np.array([0,1,0,0])
denum = np.array([1,-1,-1,1])
n1,h_n_dimpulse =sgnl.dimpulse((num,denum,1),x0=0,n=10) 
h_n_dimpulse = np.squeeze(h_n_dimpulse)

plt.figure()
plt.subplot(3,1,1), plt.stem(n,h_n_elle), plt.ylabel('$h_{elle}[n]$'), plt.title('h[n] el,residuez ve dimpulse ile çözümler'), plt.ylim(-0.3,6), plt.grid(True)
plt.subplot(3,1,2), plt.stem(n,h_n_residuez), plt.ylabel('$h_{residuez}[n]$'), plt.ylim(-0.3,6), plt.grid(True)
plt.subplot(3,1,3), plt.stem(n1,h_n_dimpulse), plt.ylabel('$h_{dimpulse}[n]$'), plt.ylim(-0.3,6), plt.grid(True)
plt.xlabel('n')

```
