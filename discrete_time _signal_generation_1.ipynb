```python
'''
AbdÃ¼lhalim YAYDEMÄ°R  
'''
# DSP environment' da numpy ve matplotlib paketleri mevcut deÄŸil onlar anaconda Ã¼zerinden eklendi 
import numpy as np 
import matplotlib.pyplot as plt           
%matplotlib notebook                     


# pi sayÄ±sÄ±nÄ± pi de sakla                             
pi = np.pi 


# gÃ¶zlemlenecek aralÄ±ÄŸÄ± belirle
n = np.arange(-10,11,1) 


# e'li ifadenin gerÃ§ek ve sanal bileÅŸenleri bulundu.
# x[n] = cos(3*pi*n/4) + jsin(3*pi*n/4)
discrete_real  = np.cos(3*pi*n/4)
discrete_imag  = np.sin(3*pi*n/4)


# gerÃ§ek ve sanal bileÅŸenler ayrÄ± ayrÄ± Ã§izdirildi
plt.subplot(1,2,1)
plt.stem(n, discrete_real, 'b-')
plt.title('x[n] iÃ§in GerÃ§ek KÄ±sÄ±m ')
plt.xlabel('n')
plt.grid(True)
plt.annotate('N = 8 Ã–RNEK', xy=(0, 1), xytext=(-17, 1.02), arrowprops=dict(facecolor='red', shrink=0.05))

plt.subplot(1,2,2)
plt.stem(n, discrete_imag, 'r-')
plt.title('x[n] iÃ§in Sanal KÄ±sÄ±m')
plt.xlabel('n')
plt.grid(True)
plt.annotate('N = 8 Ã–RNEK', xy=(6, 1), xytext=(-11, 1.02), arrowprops=dict(facecolor='blue', shrink=0.05))

plt.show()
```
